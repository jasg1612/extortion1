```python
import pandas as pd
data = pd.read_excel("output.xlsx", sheet_name="extorsion")
```

```python
####HACER LA COLUMNA UNIQUE PARA TENER SIN REPETICIONES LAS SENTENCIAS
data1=data.sentence.unique()
data_final=pd.DataFrame(data1,columns=["sentence"])
data_final
```

```python
####DETERMINAR SI ES EXTORSION O NO ACORDE A LA CATEGORIZACION A=amenaza, D=Demanda, F=familia, negocio, propiedad
d=[]
for index_sentence in range(0,len(data_final)):
  state=1
  for index_sentence2 in range(0,len(data)):
    if data_final['sentence'].values[index_sentence]==data['sentence'].values[index_sentence2]:
      if state==1:
        if data['tipo'].values[index_sentence2]=='A' or data['tipo'].values[index_sentence2]=='F':
          state=2
        if data['tipo'].values[index_sentence2]=='D':
          state=3
      if state==2:
        if data['tipo'].values[index_sentence2]=='D':
          state=4
      if state==3:
        if data['tipo'].values[index_sentence2]=='A' or data['tipo'].values[index_sentence2]=='F':
          state=4

  if state == 4:
    d.append("extorsion")
  else:
    d.append("no extorsion")
data_final.insert(1, 'PREDICTED',d)
```

```python
###CONVERTIR A EXCEL EL DATAFRAME CLASIFICADO EN CLASES EXTORSION Y NO EXTORSION
data_final.to_excel('datafinal.xlsx', sheet_name='extorsion',index=False)
```
